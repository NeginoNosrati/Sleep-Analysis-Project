**Data Cleaning and EDA **

```python
import pandas as pd

data = pd.read_csv('Sleep_Efficiency.csv')
data.info()
data.shape
```

```python
data.sample(12)
```

```python
data.isnull().sum()
```

```python
data_cleaned = data.dropna()
data_encoded = pd.get_dummies(data_cleaned, columns=['Gender'], dtype=int)

data_encoded.sample(5)
```

```python

avg_Awakanings_by_gender_data = data_encoded.groupby('Gender_Female')['Awakenings'].mean()
print(avg_Awakanings_by_gender_data)
```

**Modeling and Evaluation**

```python
from sklearn.model_selection import train_test_split

X = data_encoded[['Gender_Female']]
Y = data_encoded[['Awakenings']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=24)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, Y_train)
```

```python
perdiction = model.predict(X_test)

print("models prediction:")
print(perdiction[:5])
```

```python
score = model.score(X_test, Y_test)
print(f"score: {score}")
```
